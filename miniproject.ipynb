### 1. Loading Data
1. the first step, loading data from .csv file and create a dataframe

```python
import pandas as pd
import importlib
importlib.reload(pd)
df = pd.read_csv("https://drive.google.com/file/d/1azj6I62uNQeS07PYKYJN7WPk7vyXBZg6/view?usp=sharing")
print(df)
```

### 2. Count the average slaray by year

```python
average_salary_by_year = df.groupby('work_year')['salary_in_usd'].mean().reset_index()
```

### 3. visualize the value

```python
import matplotlib.pyplot as plt
importlib.reload(plt)
plt.figure(figsize=(10, 6))
plt.bar(average_salary_by_year['work_year'], average_salary_by_year['salary_in_usd'], color='blue')
plt.title('Average Salary by Year in USD')
plt.xlabel('Year')
plt.ylabel('Average Salary in USD')
plt.xticks(rotation=45)
plt.show()
```
